# Merge, Join and Concatenate in Pandas


```python
import pandas as pd
```

```python
df1 = pd.read_csv(r"C:\Users\subodh\OneDrive - Freedom Finance\Desktop\Subodh\IT Learning Stuffs\Learning SQL\AlexSQLTutorial\DA Portfolio Project\Python\Pandas Tutorial\LOTR.csv")
df1
```

```python
df2 = pd.read_csv(r"C:\Users\subodh\OneDrive - Freedom Finance\Desktop\Subodh\IT Learning Stuffs\Learning SQL\AlexSQLTutorial\DA Portfolio Project\Python\Pandas Tutorial\LOTR 2.csv")
df2
```

```python
df1.merge(df2)
```

```python
df1.merge(df2, how = 'inner', on = ['FellowshipID', 'FirstName'])
```

```python
df1.merge(df2, how = 'outer')
```

```python
df1.merge(df2, how = 'left')
```

```python
df1.merge(df2, how = 'right')
```

```python
df1.merge(df2, how = 'cross')
```

```python
df1.join(df2, on = 'FellowshipID', how = 'outer', lsuffix = '_Left', rsuffix = '_Right')
```

```python
df4 = df1.set_index('FellowshipID').join(df2.set_index('FellowshipID'), lsuffix = '_Left', rsuffix = '_Right', how = 'outer')
df4
```

```python
pd.concat([df1,df2])
```

```python
pd.concat([df1,df2], join = 'inner', axis = 1)
```

```python
df1.append(df2)
#the frame .append method is deprecated and will be removed from pandas in a future version. Use pandas.concat instead.
```
